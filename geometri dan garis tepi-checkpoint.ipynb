```python
# 202431095_Fawaaz Naufal 
img_f1 = cv2.imread('f1.JPG')
print(img_f1)
```

```python
# 202431095_Fawaaz Naufal
rows, cols, _ = img_f1.shape
print('img shape : ', img_f1.shape)
```

```python
# 202431095_Fawaaz Naufal
M = cv2.getRotationMatrix2D(((cols-1)/2.0, (rows-1)/2.0), 90, 1)
img_putar = cv2.warpAffine(img_f1, M, (cols, rows))
```

```python
# 202431095_Fawaaz Naufal
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
ax = axs.ravel()

axs[0].imshow(img_f1, cmap = "gray")
axs[0].set_title("Gambar Asli")

axs[1].imshow(img_putar, cmap = "gray")
axs[1].set_title("Gambar Resized")

for a in axs:
    a.axis('OFF')

plt.tight_layout()
plt.show()
```

```python
# 202431095_Fawaaz Naufal
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('novel.JPG')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

pts1 = np.float32([
    [100, 300],  
    [270, 160],  
    [300, 200],  
    [130, 340]   
])

width = 400
height = 100


pts2 = np.float32([
    [0, 0],
    [width, 0],
    [width, height],
    [0, height]
])


matrix = cv2.getPerspectiveTransform(pts1, pts2)


warped_img = cv2.warpPerspective(img_rgb, matrix, (width, height))


plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Gambar Asli + Titik Sudut")
plt.imshow(img_rgb)
plt.scatter(pts1[:, 1], pts1[:, 0], color='red', marker='o') 

plt.subplot(1, 2, 2)
plt.title("Hasil Crop Lurus (Warped)")
plt.imshow(warped_img)

plt.show()
```
